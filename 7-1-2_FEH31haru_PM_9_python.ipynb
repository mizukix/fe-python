### 7-1-2 予想問題2  
（平成31年春 基本情報技術者試験 午後 問9（C）改）

#### 〔プログラム〕

```python
InName = "sample.c"

freq = [0] * 256
infile = open(InName, "rb")
bdata = infile.read()

cnt = 0
for ch in bdata:
    cnt += 1
    freq[ch] += 1
infile.close()

print("{:10d} bytes processed\n".format(cnt))
for i in range(64):
    for ch in range(i, i + 193, 64):
        if 0x20 <= ch and ch <= 0x7E:
            print(" {:10d} {:02X} '{:c}'".format(freq[ch], ch, ch), end = "")
        else:
            print(" {:10d} {:02X}    ".format(freq[ch], ch), end = "")
    print()
```

#### 〔整列処理部〕

```python
code = []
for i in range(256):
    code.append(i)
ih = 255
while ih > 0:
    for i in range(ih):
        if freq[i] < freq[i+1]:
            code[i], code[i+1] = code[i+1], code[i]
            freq[i], freq[i+1] = freq[i+1], freq[i]
    ih -= 1
print()
for i in range(64):
    for ch in range(i, i + 193, 64):
        if 0x20 <= code[ch] and code[ch] <= 0x7E:
            print(" {:10d} {:02X} '{:c}'".format(freq[ch], code[ch], code[ch]), end = "")
        else:
            print(" {:10d} {:02X}    ".format(freq[ch], code[ch]), end = "")
    print()
```
